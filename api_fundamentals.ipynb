```python
from groq import Groq
from google.colab import userdata

# Get Groq API key from Colab Secrets
groq_key = userdata.get("GROQ_API_KEY")

# Initialize client correctly
client = Groq(api_key=groq_key)
response = client.chat.completions.create(
    model="llama-3.1-8b-instant",
    messages=[
        {"role": "user", "content": "tell me what is groq in 2 line"}
    ]
)
print(response.choices[0].message.content)
```

# Build a GroqAgent

What this agent will have

‚úî Brain ‚Üí Groq LLM

‚úî Memory ‚Üí conversation history

‚úî Behavior ‚Üí system prompt

‚úî Method ‚Üí observe ‚Üí think ‚Üí act

```python
from groq import Groq
from google.colab import userdata

class groqagent:
  def __init__(self,name):
    self.name=name
    self.memory=[]

  def think(self,task):
    self.task=task
    self.groq_key=userdata.get("GROQ_API_KEY")
    client=Groq(api_key=self.groq_key)
    res=client.chat.completions.create(
         model="llama-3.1-8b-instant",
         messages=[{"role":"user","content":task}])
    return res.choices[0].message.content

  def act(self,task):
    self.task=task
    response = self.think(task)
    # The original code called self.observe(self.task) but observe method is not defined.
    # For now, I'm removing it. If there's an intended observe functionality, it needs to be implemented.
    return response

agent=groqagent("Agent")


while True:
  task=input("<<<")
  if task=="exit":
    print("agent stopped")
    break
  response=agent.act(task)
  print(">>>",response)
```
