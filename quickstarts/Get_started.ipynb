```
from sympy import isprime
num=10
if(isprime(num)):
  print("prime")
else:
  print("not prime")
```

```
def checkPrime(n):
  if(n==2):
    return True
  else:
    for i in range(2,n):
      if(n%i==0):
        return False
      else:
        return True
num=int(input("Enter a num:"))
if(checkPrime(num)):
    print("prime")
else:
    print(" not prime")

```

**9️⃣ Decision-Based Agent**

Modify act():

input "hello" → "Hello Human"

input "status" → "Agent Active"

```
class Agent:
  def __init__(self,name):
    self.name=name
  def Act(self,task):
    if(task=="hello"):
      print("hello human")
    elif(task=="status"):
      print("agent Active")
    else:
      print("agent is working")
agent=Agent("alexa")
task=input("enter a task:")
agent.Act(task)
```
