# Build a GroqAgent

What this agent will have

‚úî Brain ‚Üí Groq LLM

‚úî Memory ‚Üí conversation history

‚úî Behavior ‚Üí system prompt

‚úî Method ‚Üí observe ‚Üí think ‚Üí act

```python
from groq import Groq
from google.colab import userdata

class GroqAgent:
  def __init__(self,name):
    self.name=name

    self.memory=[{
         "role":"system",
         "content":f"i am {self.name} assistent"}]

    #create a client once
    self.groq_key=userdata.get("GROQ_API_KEY")
    self.client=Groq(api_key=self.groq_key)



  def observe(self,task):
    #store in memeory user
    self.memory.append(
        {"role":"user",
         "content":task})

  def think(self):
      try:
         res=self.client.chat.completions.create(
             model="llama-3.1-8b-instant",
             messages=self.memory)

         reply = res.choices[0].message.content
         # store result on memeory
         self.memory.append({"role":"assistant",
                             "content":reply})
         return reply
      except Exception as e:
        return f"there is a issue{e}"


  def act(self,task):
    self.observe(task)
    return self.think()


```

**usage**

```python
agent=GroqAgent("agent")

while True:
  task=input("\n<<<")

  if task.lower()== "exit":
    print("agent stopped")
    break

  response=agent.act(task)
  print("\n>>>",response)
```
